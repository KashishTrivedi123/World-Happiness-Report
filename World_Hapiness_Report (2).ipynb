## **Display the data of top 5 east asian countries based on 'Generosity'**

```python
#Display the data of top 5 east asian countries based on 'Generosity'
east_asian_countries=df1[df1['Regional indicator']== 'East Asia']
east_asian_countries.sort_values(by='Generosity', ascending =False).head()
```

**Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)**

```python
#Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,5))
plt.xlabel('Country Name')
plt.ylabel('Ladder score')
plt.plot(east_asian_countries['Country name'][:5], east_asian_countries['Ladder score'][:5], '--bo')
plt.show()
#taken above 5 east_asian_countries already sorted by "Generosity"
```

**Build a plot that shows the variation of 'Ladder score' among the 5 south east asian countries**

```python
#Build a plot that shows the variation of 'Ladder score' among the 5 south east asian countries
south_east_asian_countries=df1[df1['Regional indicator']=='Southeast Asia']
plt.figure(figsize=(12,5))
plt.xlabel('Country Name')
plt.ylabel('Ladder score')
plt.plot(south_east_asian_countries['Country name'][:5], south_east_asian_countries['Ladder score'][:5], '--g^')
plt.show()
```

**Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'**

```python
#Create a dataframe object : df_2021 with the following countries: 
#'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. 
#Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'
df_2021 = df1[(df1["Country name"]=="China")|(df1["Country name"]=='Nepal') |(df1["Country name"]=='Bangladesh') |(df1["Country name"]=='Pakistan')|(df1["Country name"]=='Myanmar') |(df1["Country name"]=='India')|(df1["Country name"]=='Afghanistan')]
plt.figure(figsize=(12,5))
plt.xlabel('Country Name')
plt.ylabel('Logged GDP per capita')
#scatter plot below
plt.scatter(df_2021['Country name'], df_2021['Logged GDP per capita'])
plt.show()
```

```python
columns_name = ['Country name','Regional indicator','Ladder score','Logged GDP per capita','Social support',
                'Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
data = df1[columns_name].copy()
df_happy = data.rename(columns={'Ladder score':'Happiness score','Regional indicator':'Region'})
df_happy.head()
```

```python
df_happy.isnull().sum()
```

# **Happiness Vs GDP plot**






```python
# Happiness Vs GDP plot

sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (15,7)
plt.title("Happiness Vs GDP Plot")
sns.scatterplot(x = df_happy['Happiness score'], y= df_happy['Logged GDP per capita'], hue=df_happy['Region'], s =200)
plt.xlabel('Happiness Score')
plt.ylabel('GDP per Capita')
plt.legend(loc='upper left', fontsize=11)
```

# **GDP By Region**

```python
# GDP by Region

plt.rcParams['figure.facecolor']='#E0E0E0'
Region_GDP = df_happy.groupby('Region')['Logged GDP per capita'].sum() 
Region_GDP.plot.pie(autopct='%1.1f%%')
plt.title("GDP by Region")
plt.ylabel('')
```

```python
#Number of Countries in each Region
countries_region = df_happy.groupby('Region')[['Country name']].count()
countries_region
```

```python
# Correlation Map

cor = df_happy.corr(method = 'pearson')
fig, ax = plt.subplots(figsize=(10,5))
sns.heatmap(cor,annot=True,cmap='YlGnBu',ax=ax)
```

```python
# Average corruption per Region

corruption = df_happy.groupby('Region')[['Perceptions of corruption']].mean()
plt.rcParams['figure.figsize'] = (12,6)
plt.bar(corruption.index, corruption['Perceptions of corruption'],color='maroon',width = 0.6)
plt.xticks(rotation=30,ha='right')
```

Top 10 and Bottom 10 Happiest Countries

```python
# Top 10 and bottom 10 happiest countries

top_10 = df_happy.head(10)
bottom_10 = df_happy.tail(10)

fig, axes = plt.subplots(1,2,figsize=(16,6))
sns.barplot(x=top_10['Country name'], y=top_10['Healthy life expectancy'],ax=axes[0])
axes[0].set_xlabel("Country Name")
axes[0].set_ylabel("Life Expectancy")
axes[0].set_xticklabels(top_10['Country name'],rotation =45, ha='right')
axes[0].set_title("Top 10 happiest countries Life Expectancy")

sns.barplot(x=bottom_10['Country name'], y=bottom_10['Healthy life expectancy'],ax=axes[1])
axes[1].set_xlabel("Country Name")
axes[1].set_ylabel("Life Expectancy")
axes[1].set_xticklabels(bottom_10['Country name'],rotation =45, ha='right')
axes[1].set_title("Bottom 10 least happy countries Life Expectancy")
```
